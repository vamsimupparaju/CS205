```python
'''
CS205 AI PROJECT1
Venkata Vamsi K Mupparaju
UC Riverside
'''

import copy      #used to copy the array
import time      #used to calculate the run time of algorithm
import math      #used to round of decemals 

#node class to create a puzzle state node with gn,hn and currentStete array as oriperties
class node:
    def __init__(self,puzzle):
        self.gn=0
        self.hn=0
        self.currentState=puzzle
    def printPuzzle(self):
        print( "-------------")
        print( "| %i | %i | %i |" % (self.currentState[0][0],self.currentState[0][1],self.currentState[0][2]))
        print( "-------------")
        print( "| %i | %i | %i |" % (self.currentState[1][0],self.currentState[1][1],self.currentState[1][2]))
        print( "-------------")
        print( "| %i | %i | %i |" % (self.currentState[2][0],self.currentState[2][1],self.currentState[2][2]))
        print( "-------------")
      
```

```python
#main function
def main():
    inputdata=[]
    print("Welcome to CS205 Puzzle Solver")
    print("Please select\n1.Default input\n2.Enter input manually")
    mode=int(input())
    if mode==1:
        inputdata=[[1,2,3],[5,0,6],[4,7,8]]
    else:
        print("enter initial state of puzzle")
        print("enter row1 data with space seperated for each number")
        inputdata=[]
        row1=input().split(" ")
        row1=[int(x) for x in row1]
        inputdata.append(row1)
        print("enter row2 data ")
        row2=input().split(" ")
        row2=[int(x) for x in row2]
        inputdata.append(row2)
        print("enter row3 data ")
        row3=input().split(" ")
        row3=[int(x) for x in row3]
        inputdata.append(row3)

    goalState=[[1,2,3],[4,5,6],[7,8,0]]
    
    print("----------Choose Algorithm------------ \n1.Uniform cost\n2.A* with manhatten distance\n3.A* with misplaced tail")
    algoselected=int(input())
    if algoselected==1:
        print("Uniform cost search algorithm selected")
    elif algoselected==2:
        print("A* with Manhattan distance algorithm selected")
    elif algoselected==3:
        print("A* with Misplaced Tile algorithm selected")
    
    generalSearch(inputdata,algoselected,goalState)


#function toexpand the puzzle in left, right,up, down and return child nodes
def expandPuzzle(parent,visitedNodes):
    expandedNodes=[]
    parentData = parent.currentState
    #slide down
    down=copy.deepcopy(parentData)
    for i,v in enumerate(down):
        if 0 in v:
            z=v.index(0)
            if i!=2:
                down[i][z],down[i+1][z]=down[i+1][z],down[i][z]
                child=node(down)
                child.gn=parent.gn+1
                if child.currentState not in visitedNodes:
                    expandedNodes.append(child)
                break 
    #slide up
    up=copy.deepcopy(parentData)
    for i,v in enumerate(up):
        if 0 in v:
            z=v.index(0)
            if i!=0:
                up[i][z],up[i-1][z]=up[i-1][z],up[i][z]
                child=node(up)
                child.gn=parent.gn+1
                if child.currentState not in visitedNodes:
                    expandedNodes.append(child)
                break 
    #slide left
    left=copy.deepcopy(parentData)
    for i,v in enumerate(left):
        if 0 in v:
            z=v.index(0)
            if z !=0:                
                left[i][z-1],left[i][z]=left[i][z],left[i][z-1]
                child=node(left)
                child.gn=parent.gn+1
                if child.currentState not in visitedNodes:
                    expandedNodes.append(child)
                break            
    #slide right
    right=copy.deepcopy(parentData)
    for i,v in enumerate(right):
        if 0 in v:
            z=v.index(0)
            if z !=2:
                right[i][z+1],right[i][z]=right[i][z],right[i][z+1]
                child=node(right)
                child.gn=parent.gn+1
                if child.currentState not in visitedNodes:
                    expandedNodes.append(child)
                break            
 
    return expandedNodes

#function to calculate misplaced tile heuristic
def calculate_misplaced_tiles(puzzle,goalstate):
    misplaced=0
    for x in range(len(puzzle)):
        for y in range(len(puzzle[0])):
            if puzzle[x][y] !=0:
                if puzzle[x][y] !=goalstate[x][y]:
                    misplaced+=1
    return misplaced

#function to calculate manhattan distance heuristic
def calculate_manhatten_distance(puzzle,goalstate):
    distance=0
    nodesInPuzzle=[1,2,3,4,5,6,7,8]
    for num in nodesInPuzzle:
        for i in range(len(puzzle)):
            for j in range(len(puzzle[0])):
                if num==goalstate[i][j]:
                    r1,c1=i,j 
                if num==puzzle[i][j]:
                    r2,c2=i,j
        distance+=abs(r1-r2)+abs(c1-c2)
    return distance
                

#general search fUnction
def generalSearch(inputdata, algo,goalState):
    maxqueueSize=0
    nodesExpanded=0   
    queue=[]
    visited=[]
    rootNode=node(inputdata)
    rootNode.gn=0
    startTime=time.time()
    queue.append(rootNode)
    visited.append(rootNode.currentState)
    
   #infinite loop until we get goal state
    while True:
        
        queue = sorted(queue, key=lambda x: (x.hn + x.gn, x.gn))
        currentNode=queue.pop(0)
        print("Best state to expand with gn=",currentNode.gn, "and hn=",currentNode.hn,"is")
        currentNode.printPuzzle()
        if currentNode.currentState==goalState:
            print("------------------Goal State Reached---------------")
            currentNode.printPuzzle()
            print("goal state reached at depth",currentNode.gn)
            print("nodes expanded ", nodesExpanded)
            endTime=time.time()
            print("Run time:",round((endTime-startTime),2)," Seconds")
            print("Maximum queue size is",maxqueueSize)
            return
        
        childNodes=expandPuzzle(currentNode,visited)
        for child in childNodes:
            #child.gn=currentNode.gn+1
            if algo==1:
                child.hn=0
            elif algo==2:
                child.hn=calculate_manhatten_distance(child.currentState,goalState)
            elif algo==3:
                child.hn=calculate_misplaced_tiles(child.currentState,goalState)
            queue.append(child)
            visited.append(child.currentState)
            nodesExpanded +=1
            maxqueueSize=max(maxqueueSize,len(queue))

   
       
main()
```
